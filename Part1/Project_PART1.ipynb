```python
dfTA.drop("details", inplace=True,axis=1)
dfTA.drop("ratings", inplace=True,axis=1)
```

## processIndex

```python
def processIndex(x):
    x = x.split(".")
    return int(x[0])

dfTA["indexes"] = dfTA["name"].apply(processIndex)
indexes = np.unique(dfTA["indexes"])
dfTA.set_index("indexes", drop=True, inplace=True, verify_integrity=False)
```

### process features

```python
def processName(x):
    x = x.split(".")
    x = x[1][1:]
    return x
```

```python
dfTA["name"] = dfTA["name"].apply(processName)

dfTA.head(10)
```

```python
def nb_reviews(x):
    if pd.isna(x):
        return 0
    if x[0:2]=="1 ":
        return int(x[0])
    else:
        return int(x.replace(",","")[:-8])
```

```python
dfTA["nb_reviews"] = dfTA["nb_reviews"].apply(nb_reviews)

dfTA.head(1)
```

```python
def priceMinMax(x,borne):
    if pd.notna(x):
        x = x.split("-")
        min_ = int(x[0][1:])
        max_ = int(x[1][2:])
        if borne == "max":
            return max_
        else:
            return int(min_)
    else:
        return None
```

```python
dfTA["priceMin"] = dfTA["PRICE RANGE_detail"].apply(priceMinMax,borne="min")
dfTA["priceMax"] = dfTA["PRICE RANGE_detail"].apply(priceMinMax,borne="max")
dfTA = dfTA.drop(["PRICE RANGE_detail"],axis=1)
dfTA.head(1)
```
